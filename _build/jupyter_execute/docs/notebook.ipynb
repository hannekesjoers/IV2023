```python
import plotly.graph_objs as go
import plotly.express as px
import pandas as pd

df = pd.read_csv('eruptions.csv')

number = df['volcano_name'].value_counts()

fig = go.Figure(data = go.Scattergeo(
    lat = df['latitude'],
    lon = df['longitude'],
    text = 'Name: ' + df['volcano_name'],
    mode = 'markers',
    marker = dict(
            size = 5,
            symbol = 'circle'
)))
          
fig.update_layout(
    height = 600, 
    width = 1100,
    title = 'Vulkanen in de wereld',
    geo = dict(
            showland = True,
            landcolor = "rgb(250, 250, 250)",
            subunitcolor = "black",
            countrywidth = 0.5,
            subunitwidth = 0.5
        ),
)
    
fig.show()
```
